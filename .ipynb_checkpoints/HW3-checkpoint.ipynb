#####•다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오

```python
def addall1(list1):
    tmp=0
    for i in list1:
        tmp+=i
    return tmp
def addall2(list1):
    return [sum(i for i in list1)][0]
def addallodd1(list1):
    tmp=0
    for i in list1:
        if(i%2==1):
            tmp+=i
    return tmp
def addallodd2(list1):
    return [sum(i for i in list1 if i%2==1)][0]
    
print addall1([1,2,3])
print addall2([1,2,3])
print addallodd1([1,2,3,4,5])
print addallodd2([1,2,3,4,5])
```

#####1이 붙은 함수들은 그냥 평범한 방식으로 for문을 이용하여 함을 구하는 함수이고, 2가 붙은 함수는 리스트 내포를 이용하여 값을 계산하는 함수이다. 1을 구현하는 것은 큰 문제가 없으나 2를 구현할 떄 리스트 내포에 대하여 자세히 몰라 아직도 이 방법이 맞는지는 잘 모르겠다. addall2는 모든 리스트의 합을 반환하고 addallodd2는 모든 리스트 내의 홀수의 합을 sum()과 for문을 이용한 리스트 내포의 형식으로 만들어 구현하였다.

#####•사전 d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}가 주어졌을 때 다음 요구사항에 맞는 코드를 제시하시오 ◾[참고]: d.keys(), d.values()를 통해 리스트를 얻어낸 후 리스트가 지니고 있는 sort(cmp=None, key=None, reverse=False)함수를 활용하시오.
#####◾1) 키의 알파벳 오름차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.
#####◾2) 키의 알파벳 내림차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.
#####◾3) 값의 오름차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오. 
#####◾4) 값의 내림차순 순서대로 튜플 (키, 값)을 차례대로 출력하시오.




```python
d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}
keytmp=d.keys()
valuestmp=d.values()

tmp=[]
keytmp.sort()
for i in keytmp:
    tmp.append(d[i])
print zip(keytmp,tmp)

keytmp.reverse()
tmp=[]
for i in keytmp:
    tmp.append(d[i])
print zip(keytmp,tmp)

valuestmp.sort()
tmp=[]
for i in valuestmp:
    for j in keytmp:
        if(d[j]==i):
            tmp.append(j)
print zip(valuestmp,tmp)


valuestmp.reverse()
tmp=[]
for i in valuestmp:
    for j in keytmp:
        if(d[j]==i):
            tmp.append(j)
print zip(valuestmp,tmp)
```

#####먼저 사전의 key 와 value값을 따로 2개의 리스트에 저장해둔 뒤 시작하였다. key값에 대하여 출력을 할 때 에는 sort(), reverse()를 사용한 뒤 d[key]를 이용하여 value값에 접근이 가능하여 출력이 쉬웠으나 value값에 대하여 출력을 할 때 에는 d[value]로 key값에 접근이 안되어 for문을 하나 더 추가하여 원래 사전에서 key값을 찾아서 zip을 이용하여 다시 사전으로 재구성하여 출력해주었다.

#####소수를 크기 순으로 나열하면 2, 3, 5, 7, 11, 13, ... 과 같이 됩니다.

#####이 때 10,001번째의 소수를 구하세요.


```python
count=1
start=2
while(count!=10002):
    num=0
    for i in range(2,start+1):
        if(start%i==0):
            num+=1
            if(num>1):
                start+=1
                break
    if(num==1):
        start+=1
        count+=1
print start   
```

#####while문과 for문을 이용하여 현재 숫자 start가 2부터 자기 자신까지의 수로 나누었을때 1번 나누어질 경우에만 count를 증가시켰다. 하지만 이렇게 하다보니 계산이 너무 오래걸리는 것 같아 다른 여러가지를 시도해 보았지만 짝수를 모두 제외하고 시작하는 방법등을 이용하여도 걸리는 시간에 차이를 느끼지 못하여 다른 더 빠른 방법이 무엇이 있을까에 대하여 궁리중이다.

```python
def tri():
    for i in range(1,1000):
        for j in range(1,1000):
            for k in range(1,1000):
                if(i+j+k==1000 and pow(i,2)+pow(j,2)==pow(k,2)):
                    print i*j*k
                    return
tri()
```

```python
import urllib2
import string

def urlchecker(a):
    save=""
    maxnum=0
    count=0
    key1=[]
    val=[]
    source = urllib2.urlopen(a).read()
    i=0
    while(i<len(source)):
        if(source[i]=='<'):
            while(1):
                i+=1
                if(source[i]=='>'):
                    i+=1
                    break
        else:
            save+=source[i]
            i+=1
    l=string.punctuation
    key1=save.split()
    
    for i in range(len(key1)):
        for j in l:
            key1[i]=key1[i].replace(j,'')
    tmp=len(key1)
    for i in range(key1.count('')):
        key1.remove('')
            
    for i in key1:
        val.append(key1.count(i))
    i=0
    while(i<len(key1)):
        print "(",key1[i],",",val[i],")",
        i+=1

urlchecker("http://www.naver.com")
print
print
urlchecker("http://www.daum.net")
print
print
urlchecker("http://www.mytimes.com")
```
