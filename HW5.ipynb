#####3. 다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오

```python
class Counter:
    def __init__(self,v,c=1):
        self.count=v
        self.step=c
    def incr(self):
        self.count+=self.step
    def __str__(self):
        return "[Count (step:%d)] %d" % (self.step,self.count)
    def __call__(self):
        self.incr()
    def __add__(self,num):
        return Counter(self.count+num,self.step)
    def __sub__(self,num):
        return Counter(self.count-num,self.step)
    def __lt__(self, y):
        return self.count<y
    def __gt__(self, y):
        return self.count>y
    def __eq__(self, y):
        return self.count==y
    def __ne__(self, y):
        return self.count!=y
    
c = Counter(10) 
d = Counter(10, 2)
print c
print d
c.incr()
d.incr()
print c
print d
c()
d()
print c
print d
c=c+5
d=d-5
print c
print d
print c > 10
print d > 10
print c < 10
print d < 10
print c == 17
print d != 9
```

```python
import weakref
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    def __or__(self,b):
        tmp=[]
        for i in self:
            tmp.append(i)
        for i in b:
            if i not in self:
                tmp.append(i)
        return MySet(tmp)
    def __and__(self,b):
        tmp=[]
        for i in self:
            tmp.append(i)
        for i in self:
            if i not in b:
                tmp.remove(i)
        return MySet(tmp)
    def __sub__(self,b):
        tmp=[]
        for i in self:
            tmp.append(i)
        for i in b:
            if i in self:
                tmp.remove(i)
        return MySet(tmp)
    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
    
u = s | t 
print u

u = s & t 
print u

s = MySet([1, 2, 3]) 
t = MySet([3, 4, 5]) 
u = s - t 
print u 
```

```python
import urllib2
import string

class WebWordsFrequency:
    worddict={}
    def __init__(self,*args):
        self.Urls=args
        self.now=0
    def listUrls(self):
        for i in self.Urls:
            print i
    def addUrl(self,u):
        tmp=list(self.Urls)
        tmp.append(u)
        self.Urls=tuple(tmp)
    def removeUrl(self,u):
        if u in self.Urls:
            tmp=list(self.Urls)
            tmp.remove(u)
            self.Urls=tuple(tmp)
    def getWordsFrequency(self):
        for i in self.Urls:
            self.urlchecker(i)
        print self.worddict
    def urlchecker(self,a):
        save=""
        maxnum=0
        count=0
        key1=[]
        val=[]
        source = urllib2.urlopen(a).read()
        i=0
        while(i<len(source)):
            if(source[i]=='<'):
                while(1):
                    i+=1
                    if(source[i]=='>'):
                        i+=1
                        break
            else:
                save+=source[i]
                i+=1
        l=string.punctuation
        key1=save.split()
        dellist=[ 'a', 'and', 'are', 'as', 'at', 'be', 'but', 'by', 'for', 'if', 'in', 'into', 'is', 'it', 'no', 'not', 'of', 'on', 'or', 's', 'such', 't', 'that', 'the', 'their', 'then', 'there', 'these', 'they', 'this', 'to', 'was', 'will', 'with']
        for i in range(len(key1)):
            for j in l:
                key1[i]=key1[i].replace(j,'')
        for c in dellist:
            for i in range(key1.count(c)):
                key1.remove(c)
        for i in range(key1.count('')):
            key1.remove('')
        for i in key1:
            val.append(key1.count(i))
        tmpdict={}
        for i in range(len(key1)):
            tmpdict[key1[i]]=val[i]
        name=tmpdict.keys()
        value=tmpdict.values()
        for i in range(len(name)):
            if name[i] in self.worddict:
                self.worddict[name[i]]=value[i]
            else:
                self.worddict[name[i]]=value[i]
    def getMaxFreqencyWords(self):
        mostvalue=list(self.worddict.values())
        mostvalue.sort()
        for name in self.worddict.keys():
            if self.worddict[name]==mostvalue[len(mostvalue)-1]:
                print name
                
class OrderedWebWordsFrequency(WebWordsFrequency):
    listword=[]
    def getWordsFrequency(self,reverse=False):
        self.listword=[]
        for i in self.Urls:
            self.urlchecker(i)
        key = self.worddict.keys()
        value = self.worddict.values()
        for i in range(len(key)):
            self.listword.append((key[i],value[i]))
        if(reverse):
            for i in range(len(self.listword)):
                for j in range(i+1,len(self.listword)):
                    if(self.listword[i][1]>self.listword[j][1]):
                        self.listword[i],self.listword[j]=self.listword[j],self.listword[i]
        else:
            for i in range(len(self.listword)):
                for j in range(i+1,len(self.listword)):
                    if(self.listword[i][1]<self.listword[j][1]):
                        self.listword[i],self.listword[j]=self.listword[j],self.listword[i]
        print self.listword
    def __iter__(self):
        return self
    def next(self):
        try:
            tmp=self.listword[self.now]
            self.now+=1
            return tmp
        except:
            raise StopIteration
    
w1 = OrderedWebWordsFrequency('http://www.cnn.com','https://www.amazon.com') 
w1.listUrls()
print
w1.addUrl('http://www.times.com')
w1.addUrl('http://www.google.com')
w1.removeUrl('https://www.amazon.com')
w1.listUrls()
print
w1.getWordsFrequency()
print
w1.getWordsFrequency(reverse=True)
print
w1.getMaxFreqencyWords()
print
for i in w1:
    print i
```
